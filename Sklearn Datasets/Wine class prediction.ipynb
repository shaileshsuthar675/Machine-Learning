# Importing Respective Libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Loading Data

```python
from sklearn.datasets import load_wine
wine_df=load_wine()
```

# Data Visualization In Tabular Form

```python
dir(wine_df)
```

```python
df=pd.DataFrame(data=wine_df.data,columns=wine_df.feature_names)
```

```python
df.head()
```

```python
len(df.alcohol)
```

```python
df['target']=wine_df.target
```

```python
df.columns
```

```python
x=df.drop(columns='target')
y=df.target
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

# Bulit A Multinormial And Gaussian Naive Bayes Classifier 

```python
from sklearn.naive_bayes import MultinomialNB
multi_nb_model=MultinomialNB()
multi_nb_model.fit(x_train,y_train)
multi_nb_model.score(x_test,y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
gaussian_model=GaussianNB()
gaussian_model.fit(x_train,y_train)
gaussian_model.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(MultinomialNB(),x_test,y_test)
```

```python
cross_val_score(GaussianNB(),x_test,y_test)
```

```python
pred=gaussian_model.predict([[13.2, 1.78, 2.14, 11.2, 100.0, 2.65, 2.76, 0.26, 1.28, 4.38, 1.05, 3.4, 1050.0]])
```

```python
if pred[0]==0:
    print("This wine's from class_0")
elif pred[0]==1:
    print("This wine's from class_1")
else:    
    print("This wine's from class_2")
```
