#Develop the logistic regreesion model to deal with multiple dependent variables 

```python
features=['satisfaction_level','last_evaluation', 'salary_low','salary_medium']
y=hr_data.left
from sklearn.linear_model import LogisticRegression
reg_model=LogisticRegression()
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test= train_test_split(hr_data[features],hr_data['left'],test_size=0.15)
reg_model.fit(x_train,y_train)
```

```python
reg_model.score(x_test,y_test)
```
